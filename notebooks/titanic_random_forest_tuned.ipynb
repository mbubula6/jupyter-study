# Titanic Random Forest Tuned

Here I use my previous code and tune it, by deepening the use of RandomForest. I add parameters I haven't used and experiment with numbers.

## Setup

```python
#import
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# get data
data = pd.read_csv("../data/titanic_train.csv")

#define important columns
features = ['Age', 'Sex', 'Pclass', 'Fare', 'sibsp'] 

# get needed data 
x = data[features]
y = data['survived']

# split data into training datasets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Model

```python
# creating model
tuned_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=5,
    min_samples_leaf=5,
    random_state=42
)

# trees
tuned_model.fit(x_train,y_train)

# predictions
y_pred = tuned_model.predict(x_test)

# test
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
